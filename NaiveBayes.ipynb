```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("placement.csv")
print(dataset.head(3))
```

```python
sns.scatterplot(x="cgpa",y="resume_score",data=dataset,hue="placed")
plt.show()
```

```python
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=dataset["resume_score"])
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
print(gnb.score(x_test,y_test)*100)
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
print(mnb.score(x_test,y_test)*100)
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
print(bnb.score(x_test,y_test)*100)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```
